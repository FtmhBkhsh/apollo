### Imports

```python
import networkx as nx
import networkx.algorithms.isomorphism as iso
```

### Graphs initialization
#### Feel free to change this part for testing the final results.

```python
# main (code) graph
G = nx.DiGraph()
G.add_edge('c1','c2',type='generalization')
G.add_edge('c2','c3',type='association')
G.add_edge('c3','c4',type='generalization')
G.add_edge('c4','c5',type='association')
```

```python
# pattern graph
G1 = nx.DiGraph()
G1.add_edge(1,2,type='generalization')
G1.add_edge(2,3,type='association')
```

### Graphs visualizations


```python
# main (code) graph
pos=nx.spring_layout(G)
nx.draw_networkx_edge_labels(G,pos)
nx.draw(G,pos,with_labels=True)
```

```python
# pattern graph
pos=nx.spring_layout(G1)
nx.draw_networkx_edge_labels(G1,pos)
nx.draw(G1,pos,with_labels=True)
```

### Finding matched patterns

```python
import networkx.algorithms.isomorphism as iso
nm = iso.categorical_edge_match('type', 'xyz')
DiGM = iso.DiGraphMatcher(G,G1,edge_match=nm)
DiGM.subgraph_is_isomorphic()
```

```python
subgraphs = [graph for graph in DiGM.subgraph_isomorphisms_iter()]
print (subgraphs)
```
