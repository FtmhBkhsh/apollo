```python
import networkx as nx
import pickle
import matplotlib.pyplot as plt
import networkx.algorithms.isomorphism as iso
```

```python
def merge_graphs(G1,G2):
    g2_edges_list = G2.edges.data()
    for edge in g2_edges_list :
        if (G1.get_edge_data(edge[0],edge[1]) != None ):
            if (G1[edge[0]][edge[1]]['type'] != edge[2]['type']):
                G1[edge[0]][edge[1]]['type'] = G1[edge[0]][edge[1]]['type'] + '+' + edge[2]['type']
        else :
            G1.add_edge(edge[0],edge[1],type = edge[2]['type'])  
    return G1
```

```python
def match_graph(G,G1):
    nm = iso.categorical_edge_match('type', 'xyz')
    DiGM = iso.DiGraphMatcher(G,G1,edge_match=nm)
    DiGM.subgraph_is_isomorphic()
    subgraphs = [graph for graph in DiGM.subgraph_isomorphisms_iter()]
    return subgraphs
```

```python
def show_graph_with_labels(G):
    pos=nx.spring_layout(G)
    nx.draw_networkx_edge_labels(G,pos)
    nx.draw(G,pos,with_labels=True)
```

```python
planning = pickle.load(open('planning_graph.p','rb'))
```

```python
ctrl = pickle.load(open('ctrl_graph2.p','rb'))
```

# factory pattern

```python
def extract_factory(G,associat_subgraphs):
    factory_list = []
    G1 = nx.DiGraph()
    G1.add_edge(1,2,type = 'Generalization')
    general_list = match_graph(G,G1)
    for sg in associat_subgraphs:
        switch0 = 0
        switch1 = 0
        keys = [*sg]
        for i in general_list:
            j = [*i]
            if(j[1] == keys[0]):
                keys.append(j[0])
                switch0 = 1
        for i in general_list:
            j = [*i]
            if(j[1] == keys[1]):
                keys.append(j[0])
                switch1 = 1
        if(switch0 * switch1 == 1):
            factory_list.append(keys)
    return(factory_list)
        
```

```python
def show_factory():
    factory = nx.DiGraph()
    factory.add_edge(1,2,type='Association')
    ctrl_subgraphs = match_graph(ctrl,factory)
    planning_subgraphs = match_graph(planning,factory)
    f_list = extract_factory(ctrl,ctrl_subgraphs)
    g = ctrl.subgraph(f_list[0])
    show_graph_with_labels(g)
show_factory()
```

# prototype pattern

```python
def extract_prototype(G,subgraphs_list):
    pattern_list = []
    for sub in subgraphs_list:
        sub = [*sub]
        pattern_list.append(G.subgraph(sub))
    return pattern_list
```

```python
def show_prototype():
    prototype = nx.DiGraph()
    prototype.add_edge(1,2,type='Generalization')
    prototype.add_edge(2,3,type='Generalization')
    
    ctrl_subgraphs = match_graph(ctrl,prototype)
    planning_subgraphs = match_graph(planning,prototype)
    
    g = extract_prototype(planning,planning_subgraphs)[0]
    show_graph_with_labels(g)

show_prototype()
```
